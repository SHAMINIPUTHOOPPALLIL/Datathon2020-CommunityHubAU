```python
#Import the library
import pandas as pd

#Read the file
Finalschool = pd.ExcelFile('TopThreeClusters.xlsx')
SchoolProfile = pd.read_excel(Finalschool, 'Sheet1')
LGA = pd.read_excel(Finalschool, 'Sheet2')
#AEDEC =pd.ExcelFile('AEDC_atrisk.xlsx')
```

```python
##Combinig LGAs with the school information

DataCombinedNEW = pd.merge(SchoolProfile, LGA, how='outer', on=['Suburb','State'])
#print(pd.merge(SchoolProfile, LGA, left_on='Suburb', right_on='Suburb', how='left'))
      #.drop('Suburb', axis=1))
#DataCombinedNEW.head()
#DataCombinedNEW.dtypes
DataCombinedNEW.shape
```

```python
#######dropping null values
null_data = DataCombinedNEW[DataCombinedNEW.isnull().any(axis=1)]
null_data
DataCombinedNEW=DataCombinedNEW.dropna()
```

Filtering LGAs with maximum number of non-English Speaking

```python
#Casey COuncil
D1=DataCombinedNEW[DataCombinedNEW['LGA']=="Casey"]
D1.shape
```

```python
#Wyndham council
D2=DataCombinedNEW[DataCombinedNEW['LGA']=="Wyndham"]
D2.shape
```

```python
#######Final list of schools for Hub installation#####
Dview=DataCombinedNEW.iloc[:,[0,1]]
Dview
```

```python
####Saving file for further filteration with conditions such as school readiness for new hubs
NewHubSchools2020 = Dview.to_csv('NewHubSchools.csv',sep=';',encoding='utf-8')
```
